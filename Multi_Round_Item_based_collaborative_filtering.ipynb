```python
# Data processing
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings=pd.read_csv('/content/drive/Shareddrives/rs assignment1/interactions_train.csv')
# Take a look at the data
ratings.head()
```

```python
# Read in data
recipes = pd.read_csv('/content/drive/Shareddrives/rs assignment1/RAW_recipes.csv')
# Take a look at the data
recipes.head()
```

```python
recipes.rename(columns = {'id':'recipe_id'}, inplace = True)
```

```python
df = pd.merge(ratings, recipes, on='recipe_id', how='inner')
# Take a look at the data
df.head()
```

```python
new_rating = df[['user_id','recipe_id','rating','name','n_steps']].sample(frac = 0.02)
n_users=new_rating.user_id.unique().shape[0]
n_items=new_rating.recipe_id.unique().shape[0]
print('Name of users='+str(n_users)+'| Number of recipes='+ str(n_items))
train_data=new_rating
train_data.head()
```

```python
# Create user-item matrix

matrix = new_rating.pivot_table(index='recipe_id', columns='user_id', values='rating')
matrix.head()
```

```python
# Normalize user-item matrix
matrix_norm = matrix.subtract(matrix.mean(axis=1), axis = 'rows').fillna(0)
matrix_norm.head()
```

```python
from sklearn.metrics.pairwise import pairwise_distances

item_similarity=1-pairwise_distances(matrix_norm.T,metric='cosine')
print(item_similarity)
item_similarity=item_similarity.argsort()
```

```python
def input_user(mv = 246):
    liked_recipes = []
    n=int(input('Enter the no. of recommendations you want:'))
    #mv=int(input('Enter the recipe id you liked :'))
    
    print('Recipes similar to the ones you liked:')
    mv=mv+1
          #header=['movie id','movie title','release date','video release date','IMDb URL','unknown','Action','Adventure','Animation',"Children's",'Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
          #df2=pd.read_csv('u.item',sep='|',names=header,encoding='latin-1')
          #'''The movie titles are not present in the dataset u.data,it is being taken from another dataset by matching the movie ids'''
    rec_items=np.array((item_similarity[[mv],:n]))
    import matplotlib.pyplot as plt
    for mr in np.nditer(rec_items):
              #print(mr)
        for index,rows in df.iterrows():
              if(index==mr):
                  print(rows['name'])
                  print(rows['recipe_id'])
                  liked_recipes .append(rows['recipe_id']) 
                  
    return liked_recipes
```

```python
recipe_list = input_user()
while(True):

  ans = input("Do you want to update the recommendation list Y/N ?")
  if ans == 'Y':

    x = int(input("recipe you like from above "))
    if x in recipe_list:
        recipe_list.clear()
        recipe_list = input_user(mv = x)
    else:
        print("Not in a list")
  else:
    break

      
```
